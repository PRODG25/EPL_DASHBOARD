```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
#defining range of seasons to get data from (previous seasons only - until 2017/18)

years = list(range(2022, 2017, -1))
all_tables = []

standings_url = "https://fbref.com/en/comps/9/2022-2023/2022-2023-Premier-League-Stats"

#Defining a loop to get football data from all specified seasons for all teams that participated in specified seasons
for year in years:
    data = requests.get(standings_url)
    url_parts = standings_url.split('/')
    season_index = url_parts.index('comps') + 2
    season = url_parts[season_index]
    standings = pd.read_html(data.text, match="Regular season")[0]

    # Add a column for the season
    standings['Season'] = season

    # Append the table to the list
    all_tables.append(standings)

    print(standings)

    soup = BeautifulSoup(data.text)
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"

    time.sleep(1)

# Concatenate all tables into one DataFrame and clean season column
final_df = pd.concat(all_tables, ignore_index=True)
final_df['Season'] = final_df['Season'].str[:4] + '/' + final_df['Season'].str[-2:]

# Display the final DataFrame
print(final_df)

# Save the DataFrame as a CSV file
csv_path = "C:\\Users\\damia\\Desktop\\Web_Mining_Project\\end_tables.csv"
final_df.to_csv(csv_path, index=False)
```
